```python
%matplotlib inline
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df1=pd.read_csv('dim_cities.csv')
df2=pd.read_csv('dim_date.csv')
df2['date']=pd.to_datetime(df2['date'],format='%d-%b-%y')
df3=pd.read_csv('dim_plan.csv')
df4=pd.read_csv('fact_atliqo_metrics.csv')
df4['date']=pd.to_datetime(df4['date'],format='%d-%b-%y')
df4['before/after_5g']=df4['date'].apply(lambda x:'Before 5G' if x < dt.datetime(2022,6,1) else 'After 5G')
df4['Month']=df4['date'].apply(lambda x:x.month)
df5=pd.read_csv('fact_market_share.csv')
df5['date']=pd.to_datetime(df5['date'],format='%d-%b-%y')
df6=pd.read_csv('fact_plan_revenue.csv')
df6['date']=pd.to_datetime(df6['date'],format='%d-%b-%y')
```

```python
df4=df4.merge(df1,how='inner',on='city_code')
def time_period(x):
    if x==1 or x==6:
        return 1
    elif x==2 or x==7:
        return 2
    elif x==3 or x==8:
        return 3
    elif x==4 or x==9:
        return 4
df4['Period']=df4['Month'].apply(time_period)
df4
```

```python
##Market share
df_test=df5.merge(df1,how='inner',on='city_code')
df_test['Month']=df_test['date'].apply(lambda x:x.month)
df_test['before/after_5g']=df_test['date'].apply(lambda x:'Before 5G' if x < dt.datetime(2022,6,1) else 'After 5G')
df_test
```

```python
df3=df3.rename(columns={'plan':'plans'})
df6=df6.merge(df3,how='inner',on='plans')
df6=df6.merge(df1,how='inner',on='city_code')
df_test2=df6.copy()
df_test2['Month']=df_test2['date'].apply(lambda x:x.month)
df_test2['before/after_5g']=df_test2['date'].apply(lambda x:'Before 5G' if x < dt.datetime(2022,5,1) else 'After 5G')
df_test2
```
